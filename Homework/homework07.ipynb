##### 第二题 
Prove that the one-sample t-test for $H_0 : µ = µ_0$ v.s $H_1 : µ \neq µ_0$ is a GLRT under the normality assumption.

+ **Solution:**

+ Given $\Theta_0:\mu = \mu_0$ and $\Theta:\mu \neq \mu_0$, we have $$ L(\mu_0) = \prod_{i=1}^n \frac{1}{\sqrt{2\pi\hat{\sigma_0^2}}}\exp{-\frac{(\bar{x} - \mu_0)^2}{2\hat{\sigma_0^2}}} = \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\hat{\sigma_0^2}}} e^{-\frac12} $$ , $$ L(\hat{\mu}) = \prod_{i=1}^n \frac{1}{\sqrt{2\pi\hat{\sigma^2}}}\exp{-\frac{(\bar{x} - \hat{\mu})^2}{2\hat{\sigma^2}}} = \prod_{i=1}^{n}\frac{1}{\sqrt{2\pi\hat{\sigma^2}}} e^{-\frac12} $$

+ The likelihood ratio can be calculated as:



\begin{aligned}
        \lambda &= \frac{L(\mu_0)}{L(\hat{\mu})}    \\
        &= \frac{\prod_{i=1}^n \frac{1}{\sqrt{2\pi\hat{\sigma_0^2}}}\exp{-\frac{(\bar{x} - \mu_0)^2}{2\hat{\sigma_0^2}}}}{\prod_{i=1}^n \frac{1}{\sqrt{2\pi\hat{\sigma^2}}}\exp{-\frac{(\bar{x} - \hat{\mu})^2}{2\hat{\sigma^2}}}}  \\
        &= \prod_{i=1}^n \frac{\frac{1}{\sqrt{2\pi\hat{\sigma_0^2}}}}{\frac{1}{\sqrt{2\pi\hat{\sigma^2}}}}  \\
        &= [\frac{\hat{\sigma^2}}{\hat{\sigma_0^2}}]^{\frac{n}{2}}  \\
        &= [\frac{\sum_{i=1}^{n}(x_i - \bar{x})^2}{\sum_{i=1}^{n}(x_i - \mu_0)^2}]^{\frac{n}{2}}  \\
        &= \sum_{i=1}^{n} [\frac{(x_i - \bar{x})^2}{(x_i -\bar{x} + \bar{x} - \mu_0)^2}]^{\frac{n}{2}}  \\
        &= \sum_{i=1}^{n} [\frac{(x_i - \bar{x})^2}{(x_i -\bar{x})^2 + (\bar{x} - \mu_0)^2}]^{\frac{n}{2}}  \\
        &= [\frac{1}{1 + \sum_{i=1}^{n}\frac{(\bar{x} - \mu_0)^2}{(x_i - \bar{x})^2}}]^{\frac{n}{2}}  \\
        &= [\frac{1}{1 + \frac{n(\bar{x} - \mu_0)^2}{(n-1)S^2}}]^{\frac{n}{2}}  \\
        &= [\frac{1}{1 + \frac{n(\bar{x} - \mu_0)^2}{(n-1)S^2}}]^{\frac{n}{2}}  \\
        &= [\frac{1}{1 + \frac{1}{n-1}\frac{(\bar{x} - \mu_0)^2}{\hat{S^2}/n}}]^{\frac{n}{2}}  \\
        &= [\frac{1}{1 + \frac{1}{n-1}T^2}]^{\frac{n}{2}}  \\
\end{aligned}

where $ T = |\frac{(\bar{x} - \mu_0)}{S/\sqrt{n}}| \sim^{H_0} t_{n-1} $

+ Thus, given the $\lambda(T)$ is a decreasing function of $T$, the critical region can be represented by $$  c = \{\mu:0\leq \lambda \leq \lambda^* \} = \{  \mu: t = |\frac{\bar{x} - \mu_0}{S/\sqrt{n}}|>t_{n-1}(\alpha/2)  \}  $$

##### 7.4.19
The following table gives each student’s actual GMAT score minus his or her predicted score. Set up and carry out an appropriate hypothesis test. 

Use the 0.05 level of significance.

+ **Solution:**

Let $\mu$ denote the population mean of $y$, $\bar{Y}$ be the sample mean of $y$, and $S^2$ be the sample variance of $y$.

From the table, we know that $\bar{y} = \frac1n \sum_{i=1}^n y_i = 37.357142857142854 $, $S^2 = \frac{1}{n-1}\sum_{i=1}^{n}(y_i - \bar{y})^2 = 1529 $

In this example, we need to test whether students get higher GMAT scores after training, which means we need to test: $$  H_0:\mu = \mu_0 = 0 \leftrightarrow H_1:\mu>\mu_0 = 0  $$

Given $ T = \frac{\bar{Y} - \mu_0}{S/\sqrt{n}} \sim^{H_0} t_{n-1} $, we have $$ P(t = \frac{\bar{y} - \mu_0}{S/\sqrt{n}} \geq t_{n-1}(\alpha)|H_0) = \alpha $$

$$ t = \frac{\bar{y} - \mu_0}{S/\sqrt{n}} = 3.700117707700691 > 1.7613101357748562 = t_{14}(0.95) = t_{n-1}(\alpha)  $$

Thus, $H_0$ can be rejected at $\alpha = 0.05$ significance level.

In this context, we can conclue that the students get higher GMAT scores after training at $5\%$ probability of Type I error.

##### 7.4.21.

To see whether that average can be reduced, ten pipes are coated with a new plastic and buried in the same soil.

After one year, the following maximum pit depths are recorded (in inches): $(0.0039, 0.0041, 0.0038, 0.0044, 0.0040,0.0036, 0.0034, 0.0036, 0.0046,0.0036)$. 

Given that the sample standard deviation for these ten measurements is $0.00383$ inch, can it be concluded at the α = 0.05 level of significance that the plastic coating is beneficial?

+ **Solution:**

In this case, $X$ stands for the depth of the maximum pit, $\mu$ denotes the average depth of the maximum pit, $\mu_0$ denotes the hypothetical depth of the maximum pit, $\bar{X}$ denotes the sample average of depth and $S^2$ stands for the sample variance of the depth.

To test whether the average depth of the maximum pit can be reduced, we need to test:   $$ H_0:\mu = \mu_0 \text{  v.s.  } H_1:\mu<\mu_0 $$

Given that $$  Z = \frac{\bar{X} - \mu_0}{\sigma/\sqrt{n}} \sim^{H_0} N(0,1)  $$
we can compute the critical region by:
$$ P(z = \frac{\bar{X} - \mu_0}{S/\sqrt{n}} <-t_{n-1}(\alpha) |H_0) = \alpha $$

Using the sample data, we can compute the test statistic: $$ z = \frac{\bar{X} - \mu_0}{\sigma/\sqrt{n}} = -0.2476979890471315 > -1.6448536269514729 = z_{\alpha} $$

Thus, we can not reject $H_0$ at $\alpha = 0.05$ significance level. Hence, we can not conclue that the plastic coating is beneficial.

##### 第9题
Prove that the one sample variance χ2 test is a GLRT under the normality assumption

+ **Solution:**

+ Null hypothesis & Alternative Hypothesis:$$  H_0: \sigma^2 = \sigma_0^2 \text{  v.s.  } H_1: \sigma^2 \neq \sigma_0^2 $$
+ Likelihood function: $$ L(\sigma^2) = \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma^2}}\exp{-\frac{(X_i - \mu)^2}{2\sigma^2}} $$
+ Likelihood Ratio:

\begin{aligned}
    \lambda &= \frac{L(\sigma_0^2)}{L(\hat{\sigma^2})}  \\
    &= \frac{\prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma_0^2}}\exp{-\frac{(X_i - \bar{X})^2}{2\sigma_0^2}}}{\prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\hat{\sigma^2}}}\exp{-\frac{(X_i - \bar{X})^2}{2\hat{\sigma^2}}}}  \\
    &= \frac{\prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma_0^2}}\exp{-\frac{(X_i - \bar{X})^2}{2\sigma_0^2}}}{\prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\hat{\sigma^2}}}}  \\
    &= (\frac{\hat{\sigma^2}}{\sigma_0^2})^{\frac{n}{2}}\exp{-\frac{\sum_{i=1}^{n}(X_i - \bar{X})^2}{2\sigma_0^2}}  \\
    &= (\frac{\hat{\sigma^2}}{\sigma_0^2})^{\frac{n}{2}}\exp[{-\frac{n}{2}\frac{\hat{\sigma^2}}{\sigma_0^2}}]  \\
    &= (\frac{1}{n}\frac{(n-1)S^2}{\sigma_0^2})^{\frac{n}{2}}\exp[{-\frac{1}{2}\frac{(n-1)S^2}{\sigma_0^2}}] \text{, for } \hat{\sigma^2} = \frac{n-1}{n}S^2  \\
    &= (\frac{1}{n}T)^{\frac{n}{2}}\exp[{-\frac{1}{2}T}] \text{, for } \hat{\sigma^2} = \frac{n-1}{n}S^2 \text{, let } T = \frac{(n-1)S^2}{\sigma_0^2}  \\
\end{aligned}

+ Given that $\lambda(T)$ is an increasing-decreasing function of $T$, the critical region at significance level $\alpha$ is defined by: $$    P( T = \frac{(n-1)S^2}{\sigma_0^2} \geq\chi^2_{n-1}(1-\alpha/2)|H_0) +    P(\chi^2_{n-1}(\alpha/2)\geq T = \frac{(n-1)S^2}{\sigma_0^2} |H_0) = \alpha  $$    
$$ \mathrm{c} = \{T \geq \chi^2_{n-1}(1-\alpha/2) \text{ or } T\leq  \chi^2_{n-1}(\alpha/2)\} $$



##### 第10题
Suppose that the random variables $Y_1,...,Y_n$ satisfy
$$Yi = β\times x_i + \epsilon_i; \text{for }i = 1,2,..., n$$
where $x_1,...,x_n$ are non-zero fixed constants and they are not identical, $\epsilon_1,...,\epsilon_n$ are
independent and identically distributed (iid) $N(0; σ^2)$, and $σ^2$ is known. 

Hint: you first choose consider the distribution of $Y_i$, $N(β\times x_i; σ^2)$, which are independent but not identical.

1. Find the maximum likelihood estimator (MLE) of β, denoted $\hatβ_{MLE}$, and show that it is an unbiased estimator of β.
1. Find the exact distribution of the MLE of β.
2. Assume $σ^2$ is known, find the $(1 − α) ∗ 100\%$ confidence interval for β.
1. Assume $σ^2$ is known, test $H_0 : β = β_0$ vs $H_1 : β \neq β_0$ using one-sample Z test.
1. Assume $σ^2$ is unknown, find the estimator for $σ^2$

+ **Solution:**

1. Given $\epsilon^{n} \sim^{i.i.d.}N(0,\sigma^2)$ and $x^{n}$ be the constants, we know that $Y_i \sim N(\beta x_i,\sigma^2)$

    The likelihood function for $\beta$ can written as: $$ L(\beta) = \prod_{i=1}^n \frac{1}{\sqrt{2\pi\sigma^2}}\exp{-\frac{(y_i - \mu_y)^2}{2\sigma^2}} $$

    We can compute the log-likelihood function by: $$  l(\beta) = \sum_{i=1}^n -\frac12(\ln2 + \ln\pi + \ln\sigma^2)-\frac{(y_i - \beta x_i)^2}{2\sigma^2}  $$

    Differentiating by $\beta$, we have: 



\begin{align*}
        \frac{\partial l}{\partial \beta} &= \sum_{i=1}^n x_i \frac{2(y_i - \beta x_i)}{2\sigma^2} = 0\\
        \sum_{i=1}^n \frac{\beta x_i^2}{\sigma^2} &= \sum_{i=1}^n \frac{x_i y_i}{\sigma^2}  \\
        \beta &= \frac{\sum_{i=1}^{n}x_i y_i}{\sum_{i=1}^{n}x_i^2}  \\
\end{align*}

Thus, we have $$  \hat{\beta_{MLE}} = \frac{\sum_{i=1}^{n}x_i y_i}{\sum_{i=1}^{n}x_i^2}  $$
    , which is identical to the univariate OLS parameter.

To show that $\hat{\beta}_{MLE}$ is an unbiased estimator for $\beta$:



\begin{align*}
        E(\hat{\beta}_{MLE}) &= E(\frac{\sum_{i=1}^{n}x_i y_i}{\sum_{i=1}^{n}x_i^2}) \\
        &= \frac{\sum_{i=1}^{n}x_i E(y_i)}{\sum_{i=1}^{n}x_i^2} \\
        &= \frac{\sum_{i=1}^{n}x_i \beta x_i}{\sum_{i=1}^{n}x_i^2} \\
        &= \beta \frac{\sum_{i=1}^{n}x_i^2}{\sum_{i=1}^{n}x_i^2} \\
        &= \beta
\end{align*}

Thus, \hat{\beta}_{MLE} is an unbiased estimator of $\beta$

2. From $\beta = \frac{Y_i - \epsilon_i}{x_i}$ and $\epsilon \sim^{i.i.d.} N(0,\sigma^2) $, we know that $\beta \sim N[E(\hat{\beta}_{MLE}),Var(\hat{\beta}_{MLE})] = N(\beta, Var(\hat{\beta}_{MLE}))$

    To find the variance of $\hat{\beta}_{MLE}$:



\begin{align*}
        Var(\hat{\beta}_{MLE})  &= Var(\frac{\sum_{i=1}^{n}x_i y_i}{\sum_{i=1}^{n}x_i^2})   \\
        &= (\sum_{i=1}^{n}x_i^2)^{-2}Var(\sum_{i=1}^{n}x_i y_i) \\
        &= (\sum_{i=1}^{n}x_i^2)^{-2}\sum_{i=1}^{n}x_i^2 Var(y_i) \\
        &= (\sum_{i=1}^{n}x_i^2)^{-1} \sigma^2 \\
        &= \frac{\sigma^2}{\sum_{i=1}^{n}x_i^2}
\end{align*}



Thus, $$  \hat{\beta}_{MLE} \sim N(\beta,\frac{\sigma^2}{\sum_{i=1}^{n}x_i^2})  $$

3. Given that $\hat{\beta}_{MLE} \sim N(\beta,\frac{\sigma^2}{\sum_{i=1}^{n}x_i^2})$, we know that $$  P(-z_{\alpha/2}\leq \frac{\hat{\beta}_{MLE}}{\frac{\sigma^2}{\sum_{i=1}^{n}x_i^2}} \leq z_{\alpha/2})  = \alpha $$

    Thus, the $(1-\alpha)$ confidence interval for $\hat{\beta}_{MLE}$ shall be $$[\beta - z_{\alpha/2}\frac{\sigma^2}{\sum_{i=1}^{n}x_i^2},\beta + z_{\alpha/2}\frac{\sigma^2}{\sum_{i=1}^{n}x_i^2} ]$$

4. To test $H_0 : β = β_0$ against $H_1 : β \neq β_0$, we should remind that $$ \hat{\beta}_{MLE} \sim N(\beta,\frac{\sigma^2}{\sum_{i=1}^{n}x_i^2}) $$

    Thus, we have $$  P(|\frac{\hat{\beta}_{MLE} - \beta_0}{{\sigma^2}/{\sum_{i=1}^{n}x_i^2}} |>z_{\alpha/2}|H_0) = \alpha  $$

    Hence, we obtain the critical region $$ \mathrm{c} = \{  z = \frac{\hat{\beta}_{MLE} - \beta_0}{{\sigma^2}/{\sum_{i=1}^{n}x_i^2}} >z_{\alpha/2} \text{ or  }  z = \frac{\hat{\beta}_{MLE} - \beta_0}{{\sigma^2}/{\sum_{i=1}^{n}x_i^2}}<-z_{\alpha/2}\} $$
