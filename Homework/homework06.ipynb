##### 6.5.2


Let $y_1, y_2, . . . , y_{10}$ be a random sample from an exponential pdf with unknown parameter λ. 

Find the form of the GLRT for $H_0:λ=λ_0$ versus $H_1:λ\neqλ_0$. 

What integral would have to be evaluated to determine the critical value if α were equal to 0.05?

+ **Solution:**
+ Given $Y\sim \text{Exponential}(\lambda)$,$f_Y(y)= \lambda e^{-\lambda y}$.
+ So, the likelihood function of the $\lambda$ is: $$ L(\lambda) = \prod_{i=1}^n \lambda e^{-\lambda y_i} $$
+ Compute the likelihood ratio: $$ LR = \frac{L(\lambda_0)}{L(\hat{\lambda})}  $$

    Compute $\hat{\lambda}$:



\begin{aligned}
        l(\lambda) &= \sum_{i=1}^n \ln(\lambda) - \lambda y_i \\
        \text{Let }\frac{\partial l(\lambda)}{\partial \lambda} = \sum_{i=1}^n \frac{1}{\lambda} - y_i &=0   \\
        \frac{n}{\lambda} &= \sum_{i=1}^n y_i \\
        \hat{\lambda} &= \frac{n}{\sum_{i=1}^n y_i} \\
\end{aligned}


+ Derive the form of the GLRT:



\begin{aligned}
        LR &= \frac{L(\lambda_0)}{L(\hat{\lambda})} \\
        &= \frac{\prod_{i=1}^n \lambda_0 e^{\lambda_0 y_i}}{\prod_{i=1}^n \hat{\lambda}e^{\hat{\lambda}y_i}}    \\
        &= \frac{\lambda_0^n e^{\lambda_0 \sum_{i=1}^n y_i}}{\hat{\lambda}^n e^{\hat{\lambda}\sum_{i=1}^n y_i}} \\
        &= (\frac{\lambda_0 }{\hat{\lambda} })^n e^{\lambda_0 \sum_{i=1}^n y_i -n} \\
        &= (\frac{\lambda_0 }{\hat{\lambda} })^n e^{\lambda_0 \frac{n}{\hat{\lambda}} -n} \\
        &= (\frac{\lambda_0 }{\hat{\lambda} })^n e^{ n\frac{\lambda_0}{\hat{\lambda}} }e^{-n }\\
        &= (\frac{\lambda_0 }{\hat{\lambda} } e^{ \frac{\lambda_0}{\hat{\lambda}} })^n \cdot e^{-n }  \\
       T = \frac{\lambda_0 }{\hat{\lambda} } e^{ \frac{\lambda_0}{\hat{\lambda}} } &\sim^{H_0} \text{Exponential}(\frac{\lambda_0 }{\hat{\lambda} }) \\
       LR &= (T^n)e^{-n }
\end{aligned}



Obviously, $LR(T)$ is an increasing function for $T$, which means 

$$ P(0\leq LR \leq {LR}^*|H_0) = \alpha \Rightarrow P(0\leq T \leq \text{Exponential}(\frac{\lambda_0 }{\hat{\lambda} })_{\alpha}|H_0) = \alpha $$



+ Thus, when $\alpha =0.05$, an critical region can be determined by $\lambda_0$.


##### 6.5.3


Let $y_1, y_2, . . . , y_n$ be a random sample from a normal pdf with unknown mean μ and variance 1. 

Find the form of the GLRT for $H_0:μ = μ_0$ versus $H_1:μ \neq μ_0$.

+ **Solution:**
+ Given $Y\sim N(\mu,1)$, we have $f_Y(y) = \frac{1}{\sqrt{2\pi}}e^{-\frac{(x-\mu)^2}{2}} $
+ So, the likelihood function can be written as $$  L(\mu) = \prod_{i=1}^n \frac{1}{\sqrt{2\pi}}e^{-\frac{(x-\mu)^2}{2}} $$
+ Compute the likelihood ratio $$   \lambda =  \frac{L(\mu_0)}{L(\hat{\mu})}  $$
    + By MLE, $\hat{\mu} = \bar{Y} = \frac1n \sum_{i=1}^n Y_i$
+ Derive the form of the GLRT:



\begin{aligned}
        \lambda &= \frac{L(\mu_0)}{L(\hat{\mu})}    \\
        &= \frac{\prod_{i=1}^n \frac{1}{\sqrt{2\pi}}e^{-\frac{(x_i-\mu_0)^2}{2}}}{\prod_{i=1}^n \frac{1}{\sqrt{2\pi}}e^{-\frac{(x_i-\hat{\mu})^2}{2}}}\\
        &= \exp[\sum_{i=1}^n \frac{-(x_i-\mu_0)^2}{2} - \frac{-(x_i-\hat{\mu})^2}{2}] \\
        &= \exp[\sum_{i=1}^n \frac{-(x_i-\hat{\mu} + \hat{\mu}-\mu_0)^2}{2} - \frac{-(x_i-\hat{\mu})^2}{2}] \\
        &= \exp[\sum_{i=1}^n \frac{-(x_i-\hat{\mu})^2 - (\hat{\mu}-\mu_0)^2}{2} + \frac{(x_i-\hat{\mu})^2}{2}] \\
        &= \exp[\sum_{i=1}^n \frac{-(\hat{\mu}-\mu_0)^2}{2}] \\
        &= \exp[\frac{1}{2}\frac{-(\bar{Y}-\mu_0)^2}{1/{n}}] \\
\end{aligned}
    


By the CLT, $Z = \frac{(\bar{Y}-\mu_0)}{1/\sqrt{n}} = \frac{(\bar{Y}-\mu_0)}{\sigma/\sqrt{n}} \sim^{H_0} N(0,1)$



+ Determine the critical region:

    Obviously, $\lambda(Z) = \exp[\frac{1}{2}Z^2] $ is an decreasing function for $|Z|$. 
    
    Thus, we have $$  P(0\leq \lambda \leq \lambda^*|H_0) = \alpha \Rightarrow P(|Z| \geq z_{\alpha/2}|H_0) = \alpha  $$

1. Show that the critical region of the likelihood ratio test of $H_0: p = \frac12$ versus $H_1: p \neq \frac12$ can be written in the form      $$k · \ln(k) + (n − k) · \ln(n − k) ≥ λ^{∗∗} $$

##### 第7题

+ In the slides, I have shown that the Z-test for $H_0 : µ = µ_0 v.s. H_0 : µ \neq µ_0$ is a GLRT.
+ Can you show that the Z-test for $H_0 : µ = µ_0 v.s. H_0 : µ > µ_0$ is also a GLRT? 
  + The key is that you need to find the maximum value of the likelihood function over the full parameter space.

+ **Solution:**

+ Given random samples $X\sim N(\mu,\sigma^2)$, we obtain the pdf of X, $ f_X(x) = \frac{1}{\sqrt{2\pi \sigma^2}}\exp{[-\frac{(X_i-\mu)^2}{2\sigma^2}]} $
+ So, the likelihood function can be written as $$  L(\mu,\sigma^2) = \prod_{i=1}^n \frac{1}{\sqrt{2\pi \sigma^2}}\exp{[-\frac{(X_i-\mu)^2}{2\sigma^2}]} $$
+ Compute the likelihood ratio $\lambda = \frac{L(\mu_0,\sigma_0^2)}{L(\hat{\mu},\hat{\sigma^2})}$
  + $L(\mu_0,\sigma_0^2)$:
    + $ \sigma_0^2 = \frac{1}{n}\sum_{i=1}^n (X_i -\mu_0)^2 $
    + $L(\mu_0,\sigma_0^2) =  \prod_{i=1}^n\frac{1}{\sqrt{2\pi \sigma_0^2}}\exp{[-\frac{(X_i-\mu_0)^2}{2\sigma_0^2}]} = (\frac{1}{\sqrt{2\pi \sigma_0^2}})^n \exp{[-\frac{\sum_{i=1}^n(X_i-\mu_0)^2}{\frac{2}{n}\sum_{i=1}^n (X_i -\mu_0)^2}]} = (\frac{1}{\sqrt{2\pi \sigma_0^2}})^n \exp{(-\frac{n}{2})} $
  + $L(\hat{\mu},\hat{\sigma^2})$:
    + Derive the MLE of $\mu$ and $\sigma^2$:
      + Log-likelihood function: $l(\mu,\sigma^2) = \sum_{i=1}^n -\ln(\sqrt{2\pi \sigma^2}) -\frac{(X_i-\mu)^2}{2\sigma^2} = -\frac{n}{2}\ln({2\pi \sigma^2}) + \sum_{i=1}^n  -\frac{(X_i-\mu)^2}{2\sigma^2}  $
      + Differentiating by $\mu$ and $\sigma^2$:



\begin{cases}
        \frac{\partial l}{\partial \mu} &= \sum_{i=1}^n  \frac{2(X_i-\mu)}{2\sigma^2} &= 0\\
        \frac{\partial l}{\partial \sigma^2} &= -\frac{n}{2\sigma^2} + \sum_{i=1}^n  \frac{(X_i-\mu)^2}{2(\sigma^2)^2}  &= 0\\
\end{cases}

\begin{cases}
        \hat{\mu} &= \frac1n\sum_{i=1}^n X_i  \\
        \hat{\sigma^2} &= \frac1n\sum_{i=1}^n  (X_i - \hat{\mu})^2  \\
\end{cases}


+ $L(\hat{\mu},\hat{\sigma^2}) = \prod_{i=1}^n\frac{1}{\sqrt{2\pi \hat{\sigma^2}}}\exp{[-\frac{(X_i-\hat{\mu})^2}{2\hat{\sigma^2}}]} = (\frac{1}{\sqrt{2\pi \hat{\sigma^2}}})^n \exp{(-\frac{n}{2})} $
+ Simplify the likelihood ratio:



\begin{aligned}
    \lambda &= \frac{L(\mu_0,\sigma_0^2)}{L(\hat{\mu},\hat{\sigma^2})}\\
    &= \frac{(\frac{1}{\sqrt{2\pi \sigma_0^2}})^n \exp{(-\frac{n}{2})}}{(\frac{1}{\sqrt{2\pi \hat{\sigma^2}}})^n \exp{(-\frac{n}{2})}}  \\
    &= (\frac{\hat{\sigma^2}}{\sigma_0^2})^{\frac{n}{2}}  \\
    &= (\frac{\sum_{i=1}^n  (X_i - \hat{\mu})^2}{\sum_{i=1}^n (X_i -\mu_0)^2})^{\frac{n}{2}} \\
    &= (\frac{\sum_{i=1}^n  (X_i  - \hat{\mu})^2}{\sum_{i=1}^n (X_i -\hat{\mu} + \hat{\mu} -\mu_0)^2})^{\frac{n}{2}} \\
    &= (\frac{\sum_{i=1}^n (X_i  - \hat{\mu})^2}{\sum_{i=1}^n  (X_i -\hat{\mu})^2 + n(\hat{\mu} -\mu_0)^2})^{\frac{n}{2}} \\
    &= (\frac{n\hat{\sigma^2}}{n\hat{\sigma^2} + n(\bar{X} -\mu_0)^2})^{\frac{n}{2}} \\
    &= (\frac{n}{n + \frac{(\bar{X} -\mu_0)^2}{\hat{\sigma^2}/n}})^{\frac{n}{2}} \\
    &= (\frac{n}{n + Z^2})^{\frac{n}{2}}, \text{Assume that } Z = \frac{(\bar{X} -\mu_0)}{\hat{\sigma}/\sqrt{n}}, Z\sim^{H_0} N(0,1) \\
\end{aligned}


+ Derive the critical region:
  + Obviously, $\lambda(Z)$ is an decreasing function for $|Z|$. 
  + Thus, we have $$ P(0\leq \lambda \leq \lambda^*) = \alpha = P(|Z|>z_{\alpha/2}|H_0) $$
  + Similarly, when $H_1:\mu>\mu_0$, the critical region can be $$ P(Z>z_{\alpha}|H_0) = \alpha$$

##### 第9题

Let $X_1,..., X_n$ are an i.i.d. random sample of size n from normal distribution with mean µ and variance $σ^2 < 1$. We want to test $H_0 : µ = µ_0$ versus $H_1 : µ \neq µ_0$ at the significance level α.
1. Find the (1−α)100% confidence region for $µ$ such that if $µ_0$ is not in the confidence region, we can reject $H_0$ at the significance level α.
2. Find the power function $\phi(µ)$ as the function of $µ$. You can use $Φ(·)$ to denote the CDF of the standard normal distribution. You can plot the Power function in R based on this explicit form of the power function.
3. As I mentioned in the class, we can also use the R code to simulate the data and obtain a power curve. You can follow the procedure as below to do it.
   + ...

+ **Solution:**

1. Confidence interval
    + Construct the confidence interval for $\mu_0$:$$  P(L(X)\leq \mu_0 \leq U(X)) = 1-\alpha  $$
      + where $L(X)$ and $U(X)$ are statistics of the random samples $X_1,...,X_n$
    + Given that $\bar{X}\to^{P} \mu$, and $\frac{\bar{X} - \mu_0}{\sigma/\sqrt{n}}\sim^{H_0}N(0,1)$, we have $$  P(-z_{\alpha/2}\leq \frac{\bar{X} - \mu_0}{\sigma/\sqrt{n}} \leq z_{\alpha/2}) = 1-\alpha  $$
    + By simplification, we have $$ P(\bar{X} -z_{\alpha/2}\frac{\sigma}{\sqrt{n}} \leq \mu_0 \leq \bar{X} + z_{\alpha/2}\frac{\sigma}{\sqrt{n}} ) = 1-\alpha $$
    + Let $CI = [\bar{X} -z_{\alpha/2}\frac{\sigma}{\sqrt{n}},\bar{X} + z_{\alpha/2}\frac{\sigma}{\sqrt{n}}]$. We know that $P(\mu_0 \notin CI) = \alpha$, which means we can reject $H_0$ at $\alpha$ level.

2. Power function

+ By definition, the power function of $\mu$ is $\text{Power}(\mu) = 1-\beta(\mu) $
+ Computes the type II error $\beta$:



\begin{aligned}
        \beta(\mu) &= P(\text{Fail to reject }H_0|H_1)   \\
        &= P(|Z = \frac{\bar{X} - \mu_0}{\sigma/\sqrt{n}}|<z_{\alpha/2}|\mu\neq \mu_0)  \\
        &= P(-z_{\alpha/2}<Z = \frac{\bar{X} -\mu + \mu -  \mu_0}{\sigma/\sqrt{n}}<z_{\alpha/2}|\mu\neq \mu_0)  \\
        &= P(-z_{\alpha/2} - \frac{\mu -  \mu_0}{\sigma/\sqrt{n}}<Z = \frac{\bar{X} -\mu }{\sigma/\sqrt{n}}<z_{\alpha/2}- \frac{\mu -  \mu_0}{\sigma/\sqrt{n}}|\mu\neq \mu_0)  \\
        & = \Phi(z_{\alpha/2}- \frac{\mu -  \mu_0}{\sigma/\sqrt{n}}) - \Phi(-z_{\alpha/2} - \frac{\mu -  \mu_0}{\sigma/\sqrt{n}})
\end{aligned}


+ Thus, we obtain the power function: $$\text{Power}(\mu) = 1-\beta(\mu) = 1- \Phi(z_{\alpha/2}- \frac{\mu -  \mu_0}{\sigma/\sqrt{n}}) + \Phi(-z_{\alpha/2} - \frac{\mu -  \mu_0}{\sigma/\sqrt{n}}) $$
+ Assume that $\mu_0 = 0$, $\sigma = 1$, $ n = 100$, $\alpha = 0.05$. We can plot the power function.

```python
from scipy.stats import norm
import seaborn as sns
import numpy as np
alpha = 0.05
alpha_score = norm.ppf(alpha/2)
mu = np.linspace(-1,1,1000)

Z = mu*(10)
```

```python
power = 1-norm.cdf(-alpha_score - Z) + norm.cdf(alpha_score - Z)
sns.set_theme(style="darkgrid")
sns.lineplot(x = mu,y=power)
```
